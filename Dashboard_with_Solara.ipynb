```python
#Analysis crime distribution at Takoradi

def analyze_crime_distribution(file_path, location="Takoradi", crimes_of_interest=None):
    """
    Analyzes the distribution of specific crimes in a given location.
    
    Parameters:
        file_path (str): Path to the CSV file containing crime data.
        location (str): The location to filter (default: "Takoradi").
        crimes_of_interest (list): List of crimes to analyze (default: ["Vandalism", "Theft", "Homicide"]).
    
    Returns:
        pd.DataFrame: A DataFrame summarizing the counts of specified crimes in the given location.
    """
    if crimes_of_interest is None:
        crimes_of_interest = ["Vandalism", "Theft", "Homicide"]
    
    # Load the crime data
    try:
        crime_data = pd.read_csv(file_path)
    except FileNotFoundError:
        return f"Error: The file '{file_path}' was not found."
    
    # Filter data for the specified location and crimes
    filtered_data = crime_data[
        (crime_data["Location"] == location) & (crime_data["Crime"].isin(crimes_of_interest))
    ]
    
    # Count the distribution of crimes
    crime_distribution = filtered_data["Crime"].value_counts()
    
    return crime_distribution

```

```python
file_path = ghana_crime
crime_distribution = analyze_crime_distribution(file_path)

if isinstance(crime_distribution, pd.Series):
    print("Crime distribution in Takoradi for specified crimes:")
    print(crime_distribution)
else:
    print(crime_distribution)
```

```python
import folium
import folium.plugins


def crime_map(df_crime):
    latitude = 7.946527
    longitude = -1.023194

    sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)

    incidents = folium.plugins.MarkerCluster().add_to(sanfran_map)

    # loop through the dataframe and add each data point to the mark cluster
    for (
        lat,
        lng,
        label,
    ) in zip(df_crime.Latitude, df_crime.Longitude, df_crime.Crime):
        folium.Marker(
            location=[lat, lng],
            icon=None,
            popup=label,
        ).add_to(incidents)

    # show map
    display(sanfran_map)


crime_map(df_crime)
```
